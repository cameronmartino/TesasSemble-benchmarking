```python
import networkx as nx
import matplotlib.pyplot as plt
from TesasSemble.fastg_graph import fastg_to_dict
%matplotlib inline

# examples
path_g1 = 'assembly/assemble_s1/k21/assembly_graph.fastg'
path_g2 = 'assembly/assemble_s2/k21/assembly_graph.fastg'

# g1
adj_list = fastg_to_dict(path_g1)
G1 = nx.DiGraph()
G1 = nx.parse_adjlist(adj_list, create_using=G1)

# g2
adj_list = fastg_to_dict(path_g2)
G2 = nx.DiGraph()
G2 = nx.parse_adjlist(adj_list, create_using=G2)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 6));
nx.draw_networkx(G1, ax=ax,with_labels=False,node_size=5,width=2,edge_color='black')
plt.show()

fig, ax = plt.subplots(1, 1, figsize=(8, 6));
nx.draw_networkx(G2, ax=ax,with_labels=False,node_size=5,width=2,edge_color='black')
plt.show()
```
